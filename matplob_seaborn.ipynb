```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### https://github.com/mwaskom/seaborn-data
df=sns.load_dataset("titanic")
df.head()
```

```python
age = df['age'].iloc[0:5].values
age
```

```python
plt.figure(figsize=(20,5))
plt.plot(df['age'].iloc[0:5],color='#F54927',marker='*',markeredgecolor='b'
,markerfacecolor='w',linewidth=1,alpha=0.9,markersize=10)
plt.xlabel("index")
plt.ylabel("age")
plt.title("xlabel vs age")
plt.ylim(10,30)
plt.xlim(10,30)
```
