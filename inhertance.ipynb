# single level inheritance

```python
#single inharitance:
class A:
    def __init__(self,name,age):
        self.name=name 
        self.age=age 
class B(A):
    def __init__(self,name,age):
        super().__init__(name,age)
    def __str__(self):
        print(f"the name is {self.name}\nage is {self.age}")
obj=B("chandu",22)
obj.__str__()
```

# multilevel inheritance

```python
class animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age
class duck(animal):
    def __init__(self,name,age):
        super().__init__(name,age)
    def sound(self):
        print(f"the sound of {self.name} is buk buk")
    def behave(self):
        print(f"the behaviour of {self.name} is swimming")
class dog(duck):
    def __init__(self,name,age):
        super().__init__(name,age)
    def sound(self):
        print(f"the sound of {self.name} is bow bow")
    def behave(self):
        print(f"the behaviour of {self.name} is barling")
obj1=duck('junnu',4)
obj2=dog('max',5)
obj1.sound()
obj1.behave()
obj2.sound()
obj2.behave()
```

# multiple inheritance

```python
class father:
    def __init__(self,surname,name,age):
        self.surname=surname
        self.name=name
        self.age=age
class mother:
    def __init__(self,surname,name,age):
       self.surname=surname
       self.name=name
       self.age=age
class child(father,mother):
    def __init__(self,surname,name,age):
        super().__init__(surname,name,age)
    def son(self):
        print(f"the {self.name} is son of his father and mother")
obj1=child('chamanthula','chandu',22)
obj1.son()
print(obj1.surname)
print(obj1.name)
```

# hirarical inheritance

```python
class parent:
    def __init__(self,surname,name,age):
        self.surname=surname
        self.name=name
        self.age=age
class child1(parent):
    def __init__(self,surname,name,age):
        super().__init__(surname,name,age)
    def daughter(self):
        print(f"{self.name} is daughter of parent")
class child2(parent):
    def __init__(self,surname,name,age):
        super().__init__(surname,name,age)
    def son(self):
        print(f"{self.name} is son of parent")
obj1=child1("chamanthula","ankitha",19)
obj1.daughter()
obj2=child2("chamanthula","chandu",22)
obj2.son()
```
