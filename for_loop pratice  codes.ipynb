########## FOR LOOPS

```python
# 1. Print numbers from 1 to 5

for i in range(1,6):
    print(i)
```

```python
# 2. Print each letter in a string

for char in "HELLO":
    print(char)
```

```python
# 3. Sum of numbers in a list

numbers = [1,2,3,4]
total = 0
for num in numbers:
    total+=num
print(num)
```

```python
# 4. Print even numbers from 1 to 10
for i in range(1,11):
    if i%2==0:
        print(i)
```

```python
# 5. Print multiplication table of 3
for i in range(1,11):
    if i%3==0:
        print(i)
```

```python
# 6. Iterate over list with index
fruits = ["apple", "banana", "cherry"]
for i in range(len(fruits)):
    print(i)
```

```python
# 7. Print squares of numbers
for i in range(6):
    print(i**2==0)
```

```python
# 8. Loop over dictionary
student = {"name": "Alice", "age": 20}
for key in student:
    print(student[key])
```

```python
# 9. Nested loop for pairs
for i in range(3):
    for j in range(2):
        print(i,j)
```

```python
# 10. Loop over a set
colors = {"red", "blue", "green"}
for color in colors:
    print(color)
```

```python
# 11. Loop with break
for i in range(5):
    if i==3:
        break
    print(i)
```

```python
# 12. Loop with continue
for i in range(5):
    if i==2:
        continue
    print(i)
```

```python
# 13. Reverse a string
s = "hello"
for char in reversed(s):
    print(char)
```

```python
# 14. Loop through two lists
names = ["A", "B", "C"]
scores = [90, 80, 70]
for name,scores in zip(names,scores):
    print(name,scores)
```

```python
# 15. Nested list
matrix = [[1, 2], [3, 4]]
for row in matrix:
    for value in row:
        print('matrix value:',value)
```

```python
# 16. List comprehension
squares = [x**2 for x in range(5)]
print(squares)
```

```python
# 17. Loop with else
for i in range(4):
    print(i)
else:
    print("done")
```

```python
# 18. Filter vowels
for char in "education":
    if char in "aeiou":
        print(char)
```

```python
# 19. ASCII values
for char in "AB":
    print(ord(char))
```

```python
# 20. Countdown
for i in range(5,0,-1):
    print(i)
```

###############  WHILE LOOPS

```python
# 1. Count from 1 to 5
i = 1
while i<=5:
    i+=1
print(i)
```

```python
# 2. Print even numbers less than 10
i = 2
while i<10:
    i+=2
    print(i)
```

```python
# 3. Sum until 100
sum = 0
i = 1
while sum<100:
    sum+=i
    i+=1
    print(sum)
```

```python
# 4. Countdown from 5
i = 5
while i>0:
    i-=1
    print(i)
```

```python
# 5. Infinite loop with break
i = 0
while True:
    if i == 5:
        break
    i+=1
    print(i)
```

```python
# 6. Loop with continue
i = 0
while i<5:
    i+=1
    if i==3:
        continue
    print(i)
```

```python
# 7. Print powers of 2 below 100
i = 1
while i<100:
    i*=2
    print(i)
```

```python
# 8. Loop over list
lst = [1, 2, 3]
i = 0
while i < len(lst):
    i+=1
    print(i)
```

```python
# 9. Input validation
x=""
while x!="yes":
    x="yes"
    print(x)
```

```python
# 10. Factorial using while
n = 5
fact = 1
while n > 0:
    fact *= n
    n -= 1
    print(fact)
```

```python
# 11. Loop through characters
s = "data"
i = 0
while i<len(s):
    i+=1
    print("Char:", s[i])
```

```python
# 12. Loop with else
i = 0
while i<3:
    i+=1
    print(i)
else:
    print("done")
```

```python
# 13. Print digits of number
n = 123
while n>0:
    n//=10
    print(n%10)
```

```python
# 14. Nested while loop
i = 0
while i<2:
    j=0
    while j<2:
        j+=1
    i+=1
    print(i,j)
```

```python
# 15. Print odd numbers
i = 1
while i<10:
    i+=2
    print(i)
```

```python
# 16. Check prime
n=int(input())
print(n)
i=1
c=0
while i<=n:
    if n%i==0:
        c+=1
    i+=1
if c==2:
    print('prime')
else:
    print('not prime')
```

```python
# 17. Reverse number
num = 123
rev = 0
while num>0:
    rev = rev*10+num%10
    num//=10
print(rev)
```
