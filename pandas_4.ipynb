```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    'emp_id': [1,2,3,4],
    'name': ['John','Sam','Mike','Sara'],
    'dept_id': [101,102,103,104],
    'salary': [50000,60000,55000,65000]
})

departments = pd.DataFrame({
    'dept_id': [101,102,103,105],
    'department': ['HR','IT','Finance','Admin']
})
```

```python
print(employees)
```

```python
print(departments)
```

```python
df_merged = pd.merge(employees,departments,on='dept_id',how='inner')
```

```python
df_merged
```

```python
df_merged = pd.merge(employees,departments,on='dept_id',how='outer')
```

```python
df_merged
```

```python
df_merged = pd.merge(employees,departments,on='dept_id',how='left')
```

```python
df_merged
```

```python
df_merged = pd.merge(employees,departments,on='dept_id',how='right')
```

```python
df_merged
```

```python
df_merged = pd.merge(employees,departments,on='dept_id',how='right',indicator=True)
```

```python
df_merged
```

```python
df_merged = pd.merge(employees,departments,on='dept_id',how='outer',indicator=True)
```
