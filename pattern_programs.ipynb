## pratice

1. Right-angled triangle — stars

```python
n  = 4
```

```python
for i in range(1,n+1):
    print("*" * i)
```

2. Inverted right-angled triangle — stars

```python
n = 4
for i in range(n,0,-1):
    print("*" * i)
```

3. Solid square of stars

```python
n = 3
for _ in range(n):
    print("*" * n)
```

5. Number triangle (1, 12, 123, …)

```python
n = 4
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

6. Floyd’s triangle (continuous numbers)

```python
n = 4
num = 1
for i in range(1,n+1):
    for _ in range(i):
        print(num, end=" ")
        num +=1
    print()
```

7. Centered star pyramid

```python
n = 4
for i in range (1, n+1):
    spaces = " " * (n-i)
    stars = "*" * (2*i-1)
    print(spaces + stars)
```

8. Inverted centered star pyramid

```python
n = 4
for i in range (n,0,-1):
    spaces = " " * (n-i)
    stars = "*" * (2*i-1)
    print(spaces + stars)
```

9. Diamond (centered, stars)

```python
n = 4
for i in range (1, n+1):
    spaces = " " * (n-i)
    stars = "*" * (2*i-1)
    print(spaces + stars)
for i in range (n,0,-1):
    spaces = " " * (n-i)
    stars = "*" * (2*i-1)
    print(spaces + stars)
```

10. Same digit per row

```python
n = 4 
for i in range(1,n+1):
    print((str(i)+" ") * i)
```

11. Binary triangle (0 and 1)

```python
for i in range(7):
    for j in range(7):
        if  j > (i+1):
            print('*',end=' ')
    else:
        print(' ',end=' ')
    print()
```
