```python
## Calculation of time of all the functions with the help of decorator function

import time
def timer(func):
    def wrapper(*args,**kwargs):
        start = time.time

        result = func(*args,**args)

        end = time.time()
        print(f"[time] took {end-start:4f} seconds")
        return result
    return wrapper
```

```python
def my_decorator(func):
    def chandu():
        print("Before function")
        func()
        print("After function")
    return chandu
```

```python
@my_decorator
def greet():
    print("hello")
```

```python
greet()
```

# using *args and **kwargs

```python
def timer(func):
    def chandu(*args):
        print("before")
        func(*args)
        print("after")
    return chandu
```

```python
@timer
def add(a,b):
    print(a+b)
```

```python
add(2,3)
```

```python
def timer(func):
    def chandu(**kwargs):
        print("before")
        func(**kwargs)
        print("after")
    return chandu
```

```python
@timer
def add(a,b,c):
    print(a+b+c)
```

```python
add(a=1,b=2,c=3)
```
