## DROPNA

```python
import pandas as pd
import numpy as np

# Retail Dataset
data = {
    "Order_ID": ["CA-1001", "CA-1002", "CA-1003", "CA-1004"],
    "Customer_Name": ["John", np.nan, "David", "Maria"],
    "Region": ["South", "East", np.nan, "West"],
    "Category": ["Furniture", "Technology", "Office Supplies", np.nan],
    "Sales": [2500, np.nan, 1800, 3200],
    "Profit": [200, 350, np.nan, 500],
    "Discount": [0.10, np.nan, 0.20, 0.15],
    "Quantity": [2, 3, np.nan, 5]
}
```

```python
df2 = pd.DataFrame(data)
```

```python
df2
```

```python
df2.shape
```

```python
a1 = (df2.isna().sum()/len(df2))*100
```

```python
a1
```

## axis=1 means it works on columns level

```python
df2.dropna(how='any',axis=1)
```

```python
df2.dropna(how='all',axis=1)
```

## axis=0 means it works on row level

```python
df2.dropna(how='any',axis=0)
```

```python
df2.dropna(how='all',axis=0)
```

```python
(df2.dropna(axis=1,thresh=len(df2)-2))
```

## Forward Fill:(ffill)

```python
df2.ffill()
```

## Backward Fill

```python
df2.bfill()
```
