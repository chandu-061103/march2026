# Python Dictionary -

A dictionary is a collection data type in Python that stores data in key-value pairs. Think of it like a real dictionary where you look up a word (key) to find its meaning (value).

set = {value1, value2, value3}

dict = {key:value, key:value, key:value}

dict = {name: phoneNumber, lohit: 812123164 }

```python
## Chars of dicstionary

# 1. key-value paris

# Every item in a dict consists of two parts:
#  A key(identifoer)
# A value (Data associated with that key)

student = {
    "name" : "kiran" ,
    "age" : 25 ,
    "grade" : "A"
}
```
