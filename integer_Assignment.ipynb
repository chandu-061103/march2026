Question 5: Calculate the sum of digits in number 12345

```python
digits = str(12345)
sum=0
for i in digits:
    sum+=int(i)
print(sum)
```

Question 6: Check if 97 is a prime number

```python
num = 97
for i in range(2,num):
    if num%i!=0:
        print("prime number")
        break
    else:
        print("not prime number")
```

Question 7: Find the factorial of 8

```python
num = 8
result=1
for i in range(8,0,-1):
    result*=i
print(result)
```
