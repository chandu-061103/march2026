# ABSTRACTION

```python

##abstraction 
from abc import ABC, abstractmethod 

class animal(ABC):
    @abstractmethod 
    def sound(self):
        pass 
    @abstractmethod
    def colour(self):
        pass
    @abstractmethod
    def legs(self):
        pass 
class dog(animal):
    def __init__(self,name,age):
        self.name=name
        self.age=age 
    def sound(self):
        print(f"{self.name} is sounds like bow bow")
    def colour(self):
        print(f"{self.name} colour is white")
    def legs(self):
        print(f"{self.name} having 4 legs")
class tiger(animal):
    def __init__(self,name,age):
        self.name=name
        self.age=age 
    def sound(self):
        print(f"{self.name} is sounds like big")
    def colour(self):
        print(f"{self.name} colour is gold")
    def legs(self):
        print(f"{self.name} having 4 leggs")
obj1=dog("puppy",2)
obj2=tiger("king",5)
print(obj2.name)
```
