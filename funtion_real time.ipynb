# student marks system

```python
marks = []

def add_marks(mark):
    marks.append(mark)
    print("added:",{mark})

def average():
    avg = sum(marks)/len(marks)
    print("average marks:",{avg})

add_marks(80)
add_marks(90)
add_marks(100)
average()
```

# employee salary system

```python
salary = 20000

def add_bonus(bonus):
    global salary
    salary+=bonus
    print("bonus added_salary:",{salary})

def deduct_tax(tax):
    global salary
    salary-= tax
    print("tax deducted:",{salary})

add_bonus(50000)
deduct_tax(1000)
```
