```python
class Movie :
        def __init__(self,title,director,year):
            self.title = title
            self.director = director
            self.year = year
            self.rating = 0


        def display(self):
            print(f' Movie name: {self.title},relese:{self.year} ,director:{self.director}, Rating:{self.rating}')



        def update_rating(self,new_rating):
            self.rating += new_rating
            print(f'update_rating: {self.rating}')


        def is_classic(self):
            if self.year < 2000:
                print("True")
            else:
                print("False")
```

```python
m1 =Movie ("pani","jaya",2014)
m1.display()
m1.update_rating(2)
m1.display()
print(m1.is_classic())
```

```python
class Library:
    def __init__(self):
        self.books = []   # list to store Book objects

    # Add book
    def add_book(self, book):
        # Check for duplicate titles
        for b in self.books:
            if b.title == book.title:
                print("Book already exists")
                return
        
        self.books.append(book)
        print("Book added successfully")

    # Remove book by title
    def remove_book(self, title):
        for b in self.books:
            if b.title == title:
                self.books.remove(b)
                print("Book removed successfully")
                return
        
        print("Book not found")

    # Display all books
    def display_books(self):
        if not self.books:
            print("No books available")
            return
        
        for b in self.books:
            b.display()
```

```python
# Create library
lib = Library()

# Create books
b1 = book("Python Basics", "John Doe")
b2 = book("Data Structures", "Jane Smith")
b3 = book("Python Basics", "John Doe")  # duplicate

# Add books
lib.add_book(b1)
lib.add_book(b2)
lib.add_book(b3)   # should show duplicate message

# Display books
print("\nLibrary Books:")
lib.display_books()

# Remove book
lib.remove_book("Python Basics")

# Display again
print("\nAfter Removal:")
lib.display_books()

# Try removing non-existing book
lib.remove_book("C Programming")
```

```python
class library:
    def __init__(self,books):
        self.books=books

    def add_book(self,book_name,title,author):
        book = {
            'book_name':book_name,
            'title':title,
            'author':author
        }
        self.books.append(book)

    #remove title
    def remove_title(self,book_name):
        for b in self.books:
            if b[book_name]==book_name:
                self.books.remove(b)

    #display_books
    def display_books(self):
        for book in self.books:
            print(book)
```

# how to call class Attributes

```python
class Employee:
    company_name = "edukron crop"
    employee_count = 0

    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
        Employee.employee_count += 1

    @classmethod
    def change_company_name(cls,new_name):
        cls.company_name = new_name
        print(f"company name changed to : {cls.company_name}")

    @classmethod
    def get_employee_count(cls):
        return cls.employee_count
    
    def display_info(self):
        print(f"employee: {self.name},company: {Employee.company_name}")

        
```

```python
# cerating employyes

emp1 = Employee("chandu",50000)
emp2 = Employee("aravind",60000)
```
